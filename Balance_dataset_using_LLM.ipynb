```python
# Get the counts of each unique value in the 'hierarchical_label' column
hierarchical_label_counts = df['hierarchical_label'].value_counts()

# Display the hierarchical label counts
display(hierarchical_label_counts)
```

Identify classes that need downsampling (count > 200) and upsampling (count < 200) based on the hierarchical label counts.



```python
# Identify classes for downsampling (count > 200)
downsample_classes = hierarchical_label_counts[hierarchical_label_counts > 200].index.tolist()

# Identify classes for upsampling (count < 200)
upsample_classes = hierarchical_label_counts[hierarchical_label_counts < 200].index.tolist()

print("Classes for downsampling (count > 200):")
print(downsample_classes)

print("\nClasses for upsampling (count < 200):")
print(upsample_classes)
```

## Downsample majority classes

### Subtask:
Implement downsampling for classes with more than 200 records to reduce their count to 200.


**Reasoning**:
Implement downsampling for classes with more than 200 records by iterating through the downsample classes, filtering the DataFrame, sampling 200 rows for each class, and concatenating the results. Then, display the shape of the downsampled DataFrame.



```python
# Create an empty list to store the downsampled DataFrames.
downsampled_dfs = []

# Iterate through the list of downsample classes
for label in downsample_classes:
    # Filter the original DataFrame df to get only the rows belonging to that class.
    df_filtered = df[df['hierarchical_label'] == label]

    # Randomly sample 200 rows from the filtered DataFrame for the current class.
    df_sampled = df_filtered.sample(n=200, random_state=42, replace=False) # Use replace=False for sampling without replacement

    # Append the sampled DataFrame to the list created in step 1.
    downsampled_dfs.append(df_sampled)

# Concatenate all the Dataframes in the list to create a single DataFrame containing the downsampled data.
df_downsampled = pd.concat(downsampled_dfs)

# Display the shape of the df_downsampled DataFrame to confirm the downsampling.
print("Shape of the downsampled DataFrame:", df_downsampled.shape)
```

## Upsample minority classes using llm

### Subtask:
For each minority class, generate synthetic consumer complaint narratives using the Azure OpenAI GPT 4.1 model until the class reaches 200 records. This will involve prompting the LLM with existing narratives from that class.


**Reasoning**:
Define the asynchronous functions for generating synthetic narratives and upsampling minority classes, then run the main upsampling function.



**Reasoning**:
The traceback indicates that `tqdm.asyncio` is not a valid attribute. This means the progress bar functionality for asynchronous operations needs to be imported from `tqdm.asyncio` specifically. The previous import was likely just `tqdm`. I will modify the code to import `tqdm_asyncio` and use it in place of `tqdm.asyncio`.



```python
from tqdm.asyncio import tqdm_asyncio
import time

async def generate_synthetic_narrative_async(existing_narratives, retries=5, initial_delay=1):
    """Generates a synthetic consumer complaint narrative based on existing narratives with retry."""
    for attempt in range(retries):
        try:
            # Take a random sample of existing narratives to use in the prompt
            sample_size = min(len(existing_narratives), 5) # Use up to 5 examples
            sampled_narratives = random.sample(existing_narratives.tolist(), sample_size)

            prompt_text = "Generate a new consumer complaint narrative similar in style and content to the following examples:\n\n"
            for narrative in sampled_narratives:
                prompt_text += f"- {narrative}\n"
            prompt_text += "\nNew narrative:"

            messages = [
                {"role": "system", "content": "You are a helpful assistant that generates consumer complaint narratives."},
                {"role": "user", "content": prompt_text}
            ]

            response = await aclient.chat.completions.create(
                model=deployment_name,
                messages=messages,
                max_tokens=100, # Limit the length of the generated narrative
                temperature=0.7, # Allow for some creativity
                n=1 # Generate one narrative
            )
            return response.choices[0].message.content.strip()
        except Exception as e:
            print(f"Attempt {attempt + 1} failed: Error generating synthetic narrative: {e}")
            if "429" in str(e) and attempt < retries - 1:
                delay = initial_delay * (2 ** attempt) + random.uniform(0, initial_delay) # Exponential backoff with jitter
                print(f"Retrying in {delay:.2f} seconds...")
                await asyncio.sleep(delay)
            else:
                print(f"Failed to generate narrative after {retries} attempts.")
                return None
    return None


async def upsample_class_async(class_label, current_count, target_count=200):
    """Upsamples a single class to the target count by generating synthetic narratives."""
    print(f"Upsampling class: {class_label} (current count: {current_count})")
    existing_narratives = df[df['hierarchical_label'] == class_label]['consumer_complaint_narrative']
    num_synthetic_needed = target_count - current_count

    if num_synthetic_needed <= 0:
        print(f"Class {class_label} already meets or exceeds target count.")
        return df[df['hierarchical_label'] == class_label] # Return existing data if no upsampling needed

    print(f"Generating {num_synthetic_needed} synthetic narratives for {class_label}...")
    # Create a list of tasks for concurrent generation
    generation_tasks = [generate_synthetic_narrative_async(existing_narratives) for _ in range(num_synthetic_needed)]

    # Run the tasks concurrently using tqdm_asyncio for progress bar
    synthetic_narratives = await tqdm_asyncio.gather(*generation_tasks)

    # Filter out any None results from failed generations
    synthetic_narratives = [narrative for narrative in synthetic_narratives if narrative is not None]

    # Create a new DataFrame from the generated narratives
    synthetic_df = pd.DataFrame({
        'consumer_complaint_narrative': synthetic_narratives,
        'hierarchical_label': class_label
    })

    # Concatenate existing data with generated data
    upsampled_df = pd.concat([df[df['hierarchical_label'] == class_label], synthetic_df])

    print(f"Finished upsampling {class_label}. New count: {len(upsampled_df)}")
    return upsampled_df

async def upsample_minority_classes():
    """Iterates through minority classes and upsamples them."""
    upsampled_dfs_list = []
    for class_label, current_count in upsample_classes_series.items():
        upsampled_class_df = await upsample_class_async(class_label, current_count)
        upsampled_dfs_list.append(upsampled_class_df)

    # Concatenate all upsampled dataframes
    if upsampled_dfs_list:
        return pd.concat(upsampled_dfs_list)
    else:
        return pd.DataFrame(columns=['consumer_complaint_narrative', 'hierarchical_label']) # Return empty df if no upsampling

# Run the upsampling process
df_upsampled = await upsample_minority_classes()

# Display the shape of the upsampled DataFrame
print("\nShape of the upsampled DataFrame:", df_upsampled.shape)

# Display the counts of each hierarchical label in the upsampled DataFrame
print("\nHierarchical label counts in upsampled DataFrame:")
display(df_upsampled['hierarchical_label'].value_counts())
```
